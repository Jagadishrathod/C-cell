```python
A = int(input("Enter 1st input:"))
B = int(input("Enter 2nd input:"))
a = str(bin(A)[2:]) #binary conversion 
b = str(bin(B)[2:])
max_len = max(len(a), len(b))
n = max_len
#print(max_len)
a= a.zfill(max_len)
b= b.zfill(max_len)
a1=[]
for i in range(len(a)):
    a1.append(a[i])
print(a1)
b1=[]
for i in range(len(b)):
    b1.append(b[i])
print(b1)

p = []
for i in range(len(b1)-1,-1,-1):
	for j in range(len(a1)-1,-1,-1):
		sum1 = int(a[j],2) & int(b[i],2)
		p.append(sum1)
#print(p)
for i in range(0, len(p), n):
	x = p[i:i+n]
	x.reverse()
	print(x)
###################################################
su=[]
m7=p[7]
m11=p[11]
for i in range(len(p)):
    if(i==0):
        su.append(p[i])#zeroth
    elif(i==1):
        s=p[i]^p[i+3]
        c=p[i]&p[i+3]
        su.append(s)#1st
    elif(i==2):
        s1=p[i]^p[i+3]
        c1=p[i]&p[i+3]
    elif(i==3):
        s2=p[i]^p[i+3]
        c2=p[i]&p[i+3]
    elif(i==8):
        s3=p[i]^s1^c
        c3=p[i]&s1|p[i]&c|c&s1
        su.append(s3)#2nd
    elif(i==9):
        s4=p[i]^s2^c1
        c4=p[i]&s2|p[i]&c1|c1&s2
    elif(i==10):
        s5=p[i]^m7^c2
        c5=p[i]&m7|p[i]&c2|c2&m7
    elif(i==12):
        s6=p[i]^s4^c3
        c6=p[i]&s4|p[i]&c3|s4&c3
        su.append(s6)#3rd
    elif(i==13):
        s7=p[i]^s5^c4
        c7=p[i]&s5|s5&c4|p[i]&c4
        s8=s7^c6
        c8=s7&c6
        su.append(s8)#4th
    elif(i==14):
        s9=p[i]^m11^c5
        c9=p[i]&c5|m11&c5|p[i]&m11
        s10=s9^c7^c8
        c10=s9&c7|c8&c7|s9&c8
        su.append(s10)#5th
    elif(i==15):
        s11=p[i]^c9^c10
        c11=p[i]&c9|c9&c10|p[i]&c10
        su.append(s11)#6th
        su.append(c11)#7th
su.reverse()
print(su)
Result = ''.join([str(elem) for elem in su])
print(Result)
dec_res=int(Result,2)
print(dec_res)
```
